```python
# TODO
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
ds = pd.read_csv('assets/real_estate.csv', sep=';')


grupo = df_cinturonsur.groupby('level5')['price']
minimos = grupo.transform('min')
maximos = grupo.transform('max')
df_cinturonsur['price_normalized'] = (df_cinturonsur['price'] - minimos) / (maximos -minimos)

fig, ax = plt.subplots(figsize=(10, 5))

colores = ['yellow', 'red', 'black', 'blue']

for i, localidad in enumerate(Poblacion_CS):
    
    df_localidad = df_cinturonsur[df_cinturonsur['level5'] == localidad]
    ax.hist(df_localidad['price_normalized'], bins=10, color=colores[i],label= localidad, alpha=0.2)

ax.set_xlabel('Precio Normalizado')
ax.set_ylabel('Cantidad de Casas o Frecuencia')
ax.set_title('Precios Normalizados por población')
ax.legend(title='Poblaciones', bbox_to_anchor=(1.08, 1.1), loc='upper left', fontsize=12, frameon=True, ncol=2)


plt.show()
```

La frecuencia es igual al número de observaciones, en éste caso en particular, a la cantidad de casas por precio y por población. 

Si bien es cierto que las observaciones individuales por población son mayores, al momento de comparar los precios normalizados, vemos que, en conjunto, la frecuencia mayor está centrada cuando la normalización de precios tiende al precio más bajo, a las casas más baratas del mercado de cada población. Cuando los precios de las viviendas sube; en el gráfico, a partir de 0.2, el número de observaciones baja considerablemente, hay menor cantidad de casas en todas las poblaciones, a precios altos, dando como resultado una similitud en los mercados de las cuatro poblaciones de la muestra.
